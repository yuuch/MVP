```python
total_dict = {}
for col in otu_df.columns:
    total_dict[col]={}
    for i in range(8):
        total_dict[col]['level'+str(i)]={}
#print(total_dict)
        
        
```

```python
for otu in otu_df.index:
    taxo_of_otu = taxo.loc[otu][1]
    taxo_of_otu = taxo_of_otu.split(';')
    #print(taxo_of_otu)
    for sample in otu_df.columns:
        if otu_df[sample][otu] > 0:
            for i in range(len(taxo_of_otu)):
                #print(type(total_dict[sample]['level'+str(i)]))
                #print(type(i))
                total_dict[sample]['level7'][otu]=otu_df[sample][otu]
                if taxo_of_otu[i] in total_dict[sample]['level'+str(i)]:
                    total_dict[sample]['level'+str(i)][taxo_of_otu[i]]+=otu_df[sample][otu]
                else:
                    total_dict[sample]['level'+str(i)][taxo_of_otu[i]]=otu_df[sample][otu]   
        else:
            pass
```

```python
import plotly
import plotly.graph_objs as go
```

```python
def stack_bar(sample):
    mapped_level={'level0':'Kingdom','level1':'Phylum','level2':'Class','level3':'Order',
                 'level4':'Family','level5':'Genus','level6':'Species','level7':'OTU'}
    data =[]
    for level in sample:
        for ele in sample[level]:
           # print(type(sample[level][ele]))
            trace = go.Bar(
                x=[mapped_level[level]],
                y=[sample[level][ele]],
                name=ele
            )
            data.append(trace)
    layout = go.Layout(
        #showlegend=False,
        #legend=dict(x=3, y=1.2),
        autosize=False,
        height=500,
        width =1000,
        barmode='stack'
        )
    fig = go.Figure(data=data, layout=layout)
    plotly.offline.plot(fig, filename='stacked-bar.html')
```
